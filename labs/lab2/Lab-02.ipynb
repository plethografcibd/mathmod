```julia
using DifferentialEquations
using Plots
```

```julia
n=4.9
s=19.5
fi = 3*pi/4

function f(r,p,t)
    dr=r/sqrt(n^2-1)
    return dr
end

function f2(t)
    xt = tan(fi+pi)*t
    return xt
end
```

```julia
r0 = s/(n+1)
```

```julia
tetha0 = collect(LinRange(0, 2*pi, 1000))
prob = ODEProblem(f, r0, (0, 2*pi))
sol = solve(prob, saveat=tetha0)

t = collect(LinRange(0, 15, 1000))
r1=[]
tetha1=[]
for i in t
    push!(r1, sqrt(i^2+f2(i)^2))
    push!(tetha1, atan(f2(i)/i))
end
```

```julia
plot(sol, proj=:polar, label="")
plot!(tetha1, r1, proj=:polar, label="")
```

```julia
savefig("01jl.png")
```
